```python
prediction_window = pd.date_range(start = '2026-01-01', end = '2027-01-01', freq='30D')
print(prediction_window)
monthly_injection = np.random.poisson(15, len(prediction_window)) # toy data. 

monthly = pd.DataFrame(data = {'nCandidates': monthly_injection, 'month_no': np.arange(1, len(monthly_injection) + 1), 'date': prediction_window})
monthly = monthly.assign(
    next_date = monthly.date.shift(-1)
).dropna()

monthly = monthly.assign(
    nbds  = monthly.apply(
        lambda row : len(pd.date_range(row.date, row.next_date, freq = 'D')), axis = 1 # number of business days between drops
    )
)

monthly = monthly
# monthly = monthly.assign(
#     month = monthly.date.dt.month
# )

# monthly = monthly.assign(
#     first_of_month = monthly.groupby('month').date.rank(ascending=True, method = 'dense') == 1
# )

plt.plot(monthly['month_no'], monthly['nCandidates'],drawstyle = 'steps-pre')

for i, cohort_start in monthly.iterrows(): 
    print('\n')
    n_candidates = cohort_start.nCandidates
    print(f'cohort {i}: ', n_candidates)
    horizon = 0 

    for j, month_row in list(monthly.iterrows())[i:]: 
        horizon += month_row.nbds
        s_t = MC.survival_at(MC.survival_curves['client_peba'], horizon)
        remaining_candidates = n_candidates * s_t
        
        print(f'month: {month_row.next_date}, s_t = {s_t}, remaining_candidates = {int(remaining_candidates)}')

    
```

```python
# ─────────────────────────────────────────────────────────────
# 7. HEADCOUNT FORECAST PER CLIENT
# ─────────────────────────────────────────────────────────────

def forecast_client(client_id, monthly_injections, n_months=24):
    """
    Forecast active members for a given client over n_months,
    given a constant monthly injection rate.
    """
    curve = survival_curves.get(client_id, survival_curves["client_peba"])
    max_t = curve.index.max()

    forecasts = []
    for future_month in range(1, n_months + 1):
        total = 0
        for months_ago in range(1, future_month + 1):
            tenure = future_month - months_ago + 1
            s = MC.survival_at(curve, min(tenure, max_t))
            total += monthly_injections * s
        forecasts.append({"month": future_month, "active_members": round(total)})
    return pd.DataFrame(forecasts)


# Example: each client receives 20 new members/month
forecast_all = pd.concat(
    {cid: forecast_client(cid, monthly_injections=20, n_months=120)
     for cid in ["client_A", "client_B", "client_C", "client_D", "client_E"]},
    names=["client_id"]
).reset_index(level=0).reset_index(drop=True)

print("\n── 24-Month Forecast at Month 12 and 24 (20 new members/month per client) ──")
print(
    forecast_all[forecast_all["month"].isin([30, 60, 120])]
    .pivot(index="client_id", columns="month", values="active_members")
    .rename(columns={12: "month_12", 24: "month_24"})
    .to_string()
)

```
